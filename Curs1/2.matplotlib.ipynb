# Instructiunea scatter - pune punctele imprastiate


```python
lista_1 = [1, 2, 3]
lista_2 = [4, 5, 6]

plt.scatter(lista_1, lista_2);
plt.show()
```

```python
lista_1 = [1]
lista_2 = [4]

plt.scatter(lista_1, lista_2);
plt.show()
```

# Matplot poate lua si range-uri (ArrayLike)

```python
range_1 = range(4, 8)
range_2 = range(6, 10)

plt.scatter(range_1, range_2);
plt.show()
```

```python
x = [10, 10]
y = [8, 9]

plt.scatter(x, y);
```

# plt titlu


```python
x = [10, 10]
y = [8, 9]
plt.title("Acesta este titlul meu interesant")
plt.scatter(x, y);
```

# x, y

```python
x = [10, 10]
y = [8, 9]
plt.title("Acesta este titlul meu interesant")
plt.xlabel("Valoarea pe axa x")
plt.ylabel("Valoarea pe axa y")
plt.scatter(x, y);
```

# Salvarea figurii in folder

```python
x = [10, 10, 11, 20]
y = [8, 9, 10, 10]
plt.title("Acesta este titlul meu interesant")
plt.xlabel("Valoarea pe axa x")
plt.ylabel("Valoarea pe axa y")
plt.scatter(x, y);

plt.savefig("figura_mea.png")
```
