```python
import matplotlib.pyplot as plt
import csv
```

```python
countries = []
populations = []

with open("tari.csv") as freader:
    csv_reader = csv.DictReader(freader)
    print(csv_reader)

    for row in csv_reader:
        countries.append(row["country"])
        populations.append(int(row["population"]))
print("Countries", countries)
print("Populations", populations)  

       
```

```python
plt.pie(populations);
```

```python
plt.title("Populatia celor mai mari 10 tari")
plt.pie(populations);
```

```python
plt.title("Populatia celor mai mari 10 tari")
plt.pie(populations, labels=countries);
plt.savefig("10 tari.png")
```
